BOX PLOT

```python
# Create box plots for each feature grouped by species
sns.boxplot(x='variety', y='sepal.length', data=df)
plt.title('Sepal Length by Species')
plt.show()
sns.boxplot(x='variety', y='sepal.width', data=df)
plt.title('Sepal Width by Species')
plt.show()
sns.boxplot(x='variety', y='petal.length', data=df)
plt.title('Petal Length by Species')
plt.show()
sns.boxplot(x='variety', y='petal.width', data=df)
plt.title('Petal Width by Species')
plt.show()
```



1. Setosa has the shortest sepal length.
2. Versicolor Shows a slightly wider range of sepal lengths compared to Setosa.
3. Virginica has the longest sepal lengths
4. Setos has the widest sepals
5. Versicolor and Virginica have similar  sepal widths
6. Setosa has the shortest petal lengths.
7. Versicolor has a petal length greater than  Setosa.
8. Virginica has the longest petal lengths
9. Setosa has the narrowest petals
10. Versicolor shows a wider range of petal widths compared to Setosa.
11. Virginica has the widest petals.



DISTRIBUTION CHART

```python

sns.kdeplot(df['sepal.length'])
plt.title('Sepal Length Distribution')
plt.show()

sns.kdeplot(df['petal.length'])
plt.title('Petal Length Distribution')
plt.show()

sns.kdeplot(df['sepal.width'])
plt.title('Sepal width Distribution')
plt.show()

sns.kdeplot(df['petal.width'])
plt.title('Petal width Distribution')
plt.show()
```

Sepal Length: Shows a  normal distribution with a slight right skew..

Petal Length: a clear bimodal distribution, indicating two distinct groups of Iris flowers with different petal length ranges.

Sepal Width: Displays a roughly normal distribution, suggesting sepal width is  evenly distributed across the dataset.

Petal Width: shows a bimodal distribution, similar to petal length, indicating two groups with distinct petal width ranges.

SCATTERPLOT

```python
sns.scatterplot(data=df, x='sepal.length', y='sepal.width', hue='variety')
plt.title('Sepal Length vs Sepal Width, Colored by Species')
plt.show()
```

setosa has small sepal length,other two has almost same length ,but virginca has large sepal length .

SCATTER MATRIX

```python
sns.pairplot(data=df, hue='variety')
plt.show()
```

Positive correlations can be observed between:

Sepal length and petal length
Sepal length and petal width
Petal length and petal width. This suggests that Iris flowers with longer sepals tend to have longer petals

SCATTER MULTIPLE

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.scatterplot(data=df, x='sepal.length', y='sepal.width', hue='variety')
plt.title('Sepal Length vs Sepal Width')

plt.subplot(2, 2, 2)
sns.scatterplot(data=df, x='petal.length', y='petal.width', hue='variety')
plt.title('Petal Length vs Petal Width')

plt.subplot(2, 2, 3)
sns.scatterplot(data=df, x='sepal.length', y='petal.length', hue='variety')
plt.title('Sepal Length vs Petal Length')

plt.subplot(2, 2, 4)
sns.scatterplot(data=df, x='sepal.width', y='petal.width', hue='variety')
plt.title('Sepal Width vs Petal Width')

plt.tight_layout()
plt.show()
```

BUBBLE CHART

```python
plt.scatter(df['sepal.length'], df['sepal.width'], s=df['petal.length'], c=df['variety'].astype('category').cat.codes, alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Bubble Chart of Iris Dataset')
plt.show()
```

Setosa flowers  have smaller bubbles (shorter petals) and are clustered in the lower left region of the chart, indicating shorter sepal lengths and widths. Versicolor and Virginica flowers have larger bubbles (longer petals) and are clustered towards the upper right, indicating longer sepal lengths and widths.

DENSITY CHART

```python
sns.kdeplot(data=df, x='sepal.length', hue='variety')
plt.title('Density Plot of Sepal Length by Species')
plt.show()
```

Setosa irises have the shortest sepal lengths.

The distributions for Versicolor and Virginica overlap

PARALLEL CHART

```python
fig = px.parallel_coordinates(df, color=df['variety'].astype('category').cat.codes,
                              dimensions=["sepal.length", "sepal.width", "petal.length", "petal.width"],
                              title="Parallel Chart for Iris Dataset")
fig.show()
```

DEVIATION CHART

```python
std_devs = df.groupby('variety').std()

std_devs.plot(kind='bar', figsize=(10, 6))
plt.title('Standard Deviation of Features by Species')
plt.ylabel('Standard Deviation')
plt.show()
```
