```python
X = df.drop('variety', axis=1)
y = df['variety']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_train, y_train))
```

```python
pred=knn.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy}")
```

```python
import numpy as np
from collections import Counter
import pandas as pd

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))

def knn(X_train, y_train, X_test, k):
    distances = []
    for train_point in X_train.values:
        dist = euclidean_distance(X_test, train_point)
        distances.append(dist)

    k_nearest_indices = np.argsort(distances)[:k]

    k_nearest_labels = [y_train.iloc[i] for i in k_nearest_indices]

    pred_label = Counter(k_nearest_labels).most_common(1)[0][0]

    return pred_label

X_test=[6.7,3.1,4.4,1.4]
k=5
result=knn(X_train, y_train, X_test, k)
print(result)
```
