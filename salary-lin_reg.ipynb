```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
salarys = pd.read_csv('/content/drive/MyDrive/DONA_DSML_LAB/Salary_Data (1).csv')
```

```python
salarys.head()
```

```python
print(salarys.info())
```

```python
print(salarys.describe())
```

```python
sns.scatterplot(x='YearsExperience', y='Salary', data=salarys)
plt.title('Years of Experience vs. Salary')
plt.show()
```

```python
sns.histplot(salarys['Salary'], bins=10)
plt.title('Distribution of Salary')
plt.show()
```

```python
print(salarys.corr())
```

```python
X = salarys[['YearsExperience']]
y = salarys['Salary']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  # Create a LinearRegression object
regressor.fit(X_train, y_train)  # Train the model using the training data
```

```python
y_pred = regressor.predict(X_test)  # Make predictions on the test data
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print("Slope:", regressor.coef_)
print("Intercept:",regressor.intercept_)
print("accuracy",r2_score(y_test,y_pred))
```

```python
sns.scatterplot(x='YearsExperience', y='Salary', data=salarys)
plt.title('Years of Experience vs. Salary')
X = salarys[['YearsExperience']]  #
y_pred = regressor.predict(X)
plt.plot(X, y_pred, color='red', label='Fitted Line')
plt.legend()
plt.show()
```
